```python
# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M',
          '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M',
          '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded',
          '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B',
          '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B',
          '91.6B', '25.1B']

# 1
# Update Recorded Damages
def updated_damages_fun(damages):
    conversion = {"M": 1000000,
             "B": 1000000000}
    updated_damages= []
    for value in damages:
        if value == 'Damages not recorded':
            updated_damages.append(value)
        if value.find("M") != -1:
            updated_damages.append(float(value[0:value.find("M")])*conversion["M"])
        if value.find("B") != -1:
            updated_damages.append(float(value[0:value.find("B")])*conversion["B"])
    return updated_damages

# test function by updating damages
updated_damages = updated_damages_fun(damages)
print(updated_damages)   
        
```

```python
# 3
# Organizing by Year
def hurricanes_by_year(hurricanes):
    by_year= {}
    for item in hurricanes:
        year= hurricanes[item]['Year']
        if year not in by_year:
            by_year[year] = [hurricanes[item]]
        else:
            by_year[year].append(hurricanes[item])
    return by_year
    
        

# create a new dictionary of hurricanes with year and key
hurricanes_by_years= hurricanes_by_year(hurricanes)
pprint(hurricanes_by_years[1933])
```

```python
# 4
# Counting Damaged Areas
def areas(hurricanes):
    areas_dic= {}
    for hurricane in hurricanes:
        hurricane_areas= hurricanes[hurricane]['Areas Affected']
        for area in hurricane_areas:
            if area not in areas_dic:
                areas_dic[area]= 1
            else:
                areas_dic[area]+= 1
    return areas_dic

# create dictionary of areas to store the number of hurricanes involved in
hurricanes_areas_count= areas(hurricanes)
pprint(hurricanes_areas_count)
```

```python
# 5
# Calculating Maximum Hurricane Count
def area_most_hurricanes(hurricanes_areas_count):
    area_most_h= 'Antilles'
    area_count= 0
    most_area_dic= {}
    for area in hurricanes_areas_count:
        if hurricanes_areas_count[area] > area_count:
            area_most_h= area
            area_count= hurricanes_areas_count[area]
    most_area_dic[area_most_h]= area_count
    return most_area_dic
# find most frequently affected area and the number of hurricanes involved in
most_frequently_affected_area= area_most_hurricanes(hurricanes_areas_count)
pprint(most_frequently_affected_area)
```

```python
# 6
# Calculating the Deadliest Hurricane
def most_deaths(hurricanes):
    most_deaths_hurricane= 'Allen'
    death_count= 0
    most_deaths_dic= {}
    for hurricane in hurricanes:
        if hurricanes[hurricane]['Deaths'] > death_count:
            most_deaths_hurricane= hurricane
            death_count= hurricanes[hurricane]['Deaths']
        
    most_deaths_dic[most_deaths_hurricane]= death_count 
    return most_deaths_dic          
            


# find highest mortality hurricane and the number of deaths
highest_mortality_hurricane= most_deaths(hurricanes)
print(highest_mortality_hurricane)
```

```python
# 7
# Rating Hurricanes by Mortality
def hurricanes_morality_rate(hurricanes):
    mora_rate= {0:[], 1:[], 2:[], 3:[], 4:[], 5:[]}
    for hurrican in hurricanes:
        mora_count= hurricanes[hurrican]['Deaths']
        if mora_count == 0:
            mora_rate[0].append(hurrican)
        elif  mora_count > 0 and mora_count <= 100:
            mora_rate[1].append(hurrican)
        elif mora_count > 100 and mora_count <= 500:
            mora_rate[2].append(hurrican)
        elif mora_count > 500 and mora_count <= 1000:
            mora_rate[3].append(hurrican)
        elif mora_count > 1000 and mora_count <= 10000:
            mora_rate[4].append(hurrican)
        elif mora_count > 10000:
            mora_rate[5].append(hurrican)
    return mora_rate
        

# categorize hurricanes in new dictionary with mortality severity as key
morality_rate_huricanes= hurricanes_morality_rate(hurricanes)
pprint(morality_rate_huricanes)
```

```python
# 8
# Calculating Hurricane Maximum Damage
def damage_conviserion_fun(hurricanes):
    conversion = {"M": 1000000, "B": 1000000000}
    updated_hurricanes_dmg= hurricanes
    for hurricane in hurricanes:
        dmg= hurricanes[hurricane]['Damage']
        if isinstance(dmg, str):
            if dmg == 'Damages not recorded':
                updated_hurricanes_dmg[hurricane]['Damage']= dmg
            if dmg.find("M") != -1:
                updated_hurricanes_dmg[hurricane]['Damage']= float(dmg[0:dmg.find("M")])*conversion["M"]
            if dmg.find("B") != -1:
                updated_hurricanes_dmg[hurricane]['Damage']= float(dmg[0:dmg.find("B")])*conversion["B"]
    return updated_hurricanes_dmg
new_hurricanes_dmg= damage_conviserion_fun(hurricanes)

# find highest damage inducing hurricane and its total cost
def highest_dmg_hurricane(new_hurricanes_dmg):
    highest_dmg= None
    hur_count= 0
    highest_dmg_dic= {}
    for hurricane in new_hurricanes_dmg:
        dmg= new_hurricanes_dmg[hurricane]['Damage']
        if dmg == 'Damages not recorded':
            continue
        if dmg > hur_count:
            highest_dmg= hurricane
            hur_count= dmg
    highest_dmg_dic[highest_dmg]= hur_count
    return highest_dmg_dic
highest_damage_hurricane= highest_dmg_hurricane(new_hurricanes_dmg)
print(highest_damage_hurricane)
            

```

```python
# 9
# Rating Hurricanes by Damage
def hurricanes_dmg_rate(new_hurricanes_dmg):
    damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}
    dmg_rate_dic= {0:[], 1:[], 2:[], 3:[], 4:[], 5:[]}
    for hurricane in hurricanes:
        dmg = hurricanes[hurricane]['Damage']
        if dmg == 'Damages not recorded':
            dmg_rate_dic[0].append(hurricanes[hurricane])            
        elif dmg == damage_scale[0]:
            dmg_rate_dic[0].append(hurricanes[hurricane])
        elif dmg > damage_scale[0] and dmg < damage_scale[1]:
            dmg_rate_dic[1].append(hurricanes[hurricane])
        elif dmg > damage_scale[1] and dmg < damage_scale[2]:
            dmg_rate_dic[2].append(hurricanes[hurricane])
        elif dmg > damage_scale[2] and dmg < damage_scale[3]:
            dmg_rate_dic[3].append(hurricanes[hurricane])
        elif dmg > damage_scale[3] and dmg < damage_scale[4]:
            dmg_rate_dic[4].append(hurricanes[hurricane])
        elif dmg > damage_scale[4]:
            dmg_rate_dic[5].append(hurricanes[hurricane])
    return dmg_rate_dic

            


# categorize hurricanes in new dictionary with damage severity as key
Hurricanes_rated_by_damaged= hurricanes_dmg_rate(new_hurricanes_dmg)
pprint(Hurricanes_rated_by_damaged)
```
