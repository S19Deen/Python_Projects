```python
Genres: ['Action', 'Adventure', 'Avant Garde', 'Award Winning', 'Boys Love', 'Comedy', 'Drama', 'Ecchi', 'Erotica', 'Fantasy', 'Girls Love', 'Gourmet',
         'Hentai', 'Horror', 'Mystery', 'Romance', 'Sci-Fi', 'Slice of Life', 'Sports', 'Supernatural', 'Suspense']
Types: ['Movie', 'Music', 'ONA', 'OVA', 'Special', 'TV']
Sources: ['4-koma manga', 'Book', 'Card game', 'Game', 'Light novel', 'Manga', 'Mixed media', 'Music', 'Novel', 'Original', 'Other',
          'Picture book', 'Radio', 'Visual novel', 'Web manga', 'Web novel']
Ratings: ['G - All Ages', 'PG - Children', 'PG-13 - Teens 13 or older', 'R - 17+ (violence & profanity)', 'R+ - Mild Nudity', 'Rx - Hentai', 'UNKNOWN']
Statuses: ['Currently Airing', 'Finished Airing', 'Not yet aired']
          
          

          
```

```python
anime['Score'] = anime['Score'].replace('UNKNOWN', 0.0)
anime['Score'] = pd.to_numeric(anime['Score'], errors='coerce') 
```

```python
def anime_recommendation_fun(data, score, genres, types, source, rating, status):
    anime_id_rec = []  # List to store recommended anime IDs
    rec_anime_dic = {}  # Dictionary to store recommended animes
    
    for _, row in data.iterrows():
        # Check if each parameter is provided; if not, treat it as "all possible results"
        if (
            (score is None or float(row['Score']) >= score)
            and (not genres or any(genre in row['Genres'].split(',') for genre in genres))  # If genres is empty, no filtering on genres
            and (not types or row['Type'] in types)  # If types is empty, no filtering on types
            and (not source or row['Source'] in source)  # If source is empty, no filtering on source
            and (rating is None or row['Rating'] == rating)  # If rating is None, no filtering on rating
            and (status is None or row['Status'] == status)  # If status is None, no filtering on status
        ):
            anime_id_rec.append(row['anime_id'])  # Add matching anime_id to the list
    
    # Create the dictionary of recommended animes
    for _, row in data.iterrows():
        if row['anime_id'] in anime_id_rec:  # If anime_id matches
            rec_anime_dic[row['anime_id']] = row.to_dict()  # Store the row as a dictionary with anime_id as key
    
    return rec_anime_dic
```

```python
recommendations = anime_recommendation_fun(anime, 8.0, ['Action'], ['TV'], ['Original'], 'R - 17+ (violence & profanity)', 'Finished Airing')
pp(recommendations)  # Print the recommended anime
```
